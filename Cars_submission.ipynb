```python
# collecting column values as np arrays
price  = np.array(data['Price'])
price = price.reshape(len(price),1)

make = np.array(data['Make'])
make = make.reshape(len(make),1)

model = np.array(data['Model'])
model = model.reshape(len(model),1)

cylinder = np.array(data['Cylinder'])
cylinder = cylinder.reshape(len(cylinder),1)

liter = np.array(data['Liter'])
liter = liter.reshape(len(liter),1)

door = np.array(data['Doors'])
door = door.reshape(len(door),1)

cruise = np.array(data['Cruise'])
cruise = cruise.reshape(len(cruise),1)

sound = np.array(data['Sound'])
sound = sound.reshape(len(sound),1)

leather = np.array(data['Leather'])
leather = leather.reshape(len(leather),1)

trim  = np.array(data['Trim'])
trim = trim.reshape(len(trim),1)

types  = np.array(data['Type'])
types = types.reshape(len(types),1)
```

```python
#### Making a big model with all variables, including some transformations
### first encode make, model, type, trim, cylinder, liter, doors, mileage, cruise, leather

#makes_encoded is the models data in an encoded matrix
enc = preprocessing.OneHotEncoder()
enc.fit(make)
makes_encoded = enc.transform(make).toarray()
makes_encoded.shape

#same for models
#models_encoded is the models data in an encoded matrix
enc.fit(model)
models_encoded = enc.transform(model).toarray()
models_encoded.shape

#same for trims
#trims_encoded is the models data in an encoded matrix
enc.fit(trim)
trims_encoded = enc.transform(trim).toarray()
trims_encoded.shape

#same for types
#types_encoded is the types data in an encoded matrix
enc.fit(types)
types_encoded = enc.transform(types).toarray()
types_encoded.shape

#same for cylinders
#cylinders_encoded is the cylinder data in an encoded matrix
enc.fit(cylinder)
cylinders_encoded = enc.transform(cylinder).toarray()
cylinders_encoded.shape

#same for liter
#liters_encoded is the models data in an encoded matrix
enc.fit(liter)
liters_encoded = enc.transform(liter).toarray()
liters_encoded.shape

#same for doors
#doors_encoded is the models data in an encoded matrix
enc.fit(door)
doors_encoded = enc.transform(door).toarray()
doors_encoded.shape

#same for cruise
#cruises_encoded is the models data in an encoded matrix
enc.fit(cruise)
cruises_encoded = enc.transform(cruise).toarray()
cruises_encoded.shape

#same for leather
#leathers_encoded is the models data in an encoded matrix
enc.fit(leather)
leathers_encoded = enc.transform(leather).toarray()
leathers_encoded.shape

#same for sound
#sounds_encoded is the models data in an encoded matrix
enc.fit(sound)
sounds_encoded = enc.transform(sound).toarray()
sounds_encoded.shape
```

```python
# making a large dataframe with all variables

X = pd.DataFrame()
X['mileage'] = data['Mileage']

X = X.join(pd.DataFrame(makes_encoded), rsuffix='_')
print(X.shape)
X = X.join(pd.DataFrame(models_encoded), rsuffix='_')
print(X.shape)
X = X.join(pd.DataFrame(trims_encoded), rsuffix='_')
print(X.shape)
X = X.join(pd.DataFrame(types_encoded), rsuffix='_')
print(X.shape)
X = X.join(pd.DataFrame(cylinders_encoded), rsuffix='_')
print(X.shape)
X = X.join(pd.DataFrame(liters_encoded), rsuffix='_')
print(X.shape)
X = X.join(pd.DataFrame(doors_encoded), rsuffix='_')
print(X.shape)
X = X.join(pd.DataFrame(sounds_encoded), rsuffix='_')
print(X.shape)
X = X.join(pd.DataFrame(cruises_encoded), rsuffix='_')
print(X.shape)
X = X.join(pd.DataFrame(leathers_encoded), rsuffix='_')
print(X.shape)
```
